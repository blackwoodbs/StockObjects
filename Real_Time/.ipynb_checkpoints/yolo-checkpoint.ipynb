```python
driver = webdriver.Chrome(PATH)
driver.get(aapl)

while True:
    driver.save_screenshot('ss.png')
    img = cv2.imread('ss.png')

    height, width, _ = img.shape
    
#     print(f'height = {height}, width = {width}')

    blob = cv2.dnn.blobFromImage(img, 1/255, (608, 608), (0, 0, 0), swapRB=True, crop=False) #608 > 416

    net.setInput(blob)

    output_layers_names = net.getUnconnectedOutLayersNames()
    layer_outputs = net.forward(output_layers_names)

    boxes = []
    confidences = []
    class_ids = []

#     for output in layer_outputs:
#     for detection in layer_outputs[0,0]:

#         score = float(detection[2])
#         if score > 0.2:

#             left = detection[3] * cols
#             top = detection[4] * rows
#             right = detection[5] * cols
#             bottom = detection[6] * rows

#             #draw a red rectangle around detected objects
#             cv2.rectangle(img, (int(left), int(top)), (int(right), int(bottom)), (0, 0, 255), thickness=2)

    
    
    
    
    
    for output in layer_outputs:
        for detection in output:
            
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
#             if confidence > 0.50:
            if float(detection[2]) > 0.9:  
#############################
                left = detection[3] * width
                top = detection[4] * height
                right = detection[5] * width
                bottom = detection[6] * height
                
                print(f'left:{left} top:{top} right:{right} bottom:{bottom}')

                #draw a red rectangle around detected objects
                cv2.rectangle(img, (int(left), int(top)), (int(right), int(bottom)), (0, 0, 255), thickness=1)
############################
                
                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)

                # corrections to find top left cordinates of detection
                x = int(center_x - 0.5*w)
                y = int(center_y - 0.5*h)

                boxes.append([x, y, w, h])
                confidences.append((float(confidence)))
                class_ids.append(class_id)

    indexes = np.array(cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4))
    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0, 255, size=(len(boxes), 3))

#     print(len(indexes))# = ~2800
#     print(len(boxes))
    
#     print(indexes.flatten()[0:2])
#     print(boxes[indexes.flatten()[0]])
#     print(boxes[indexes.flatten()[1]])
    
#     print(classes)
#     print(class_ids)
    
    for i in indexes.flatten():
        if str(classes[class_ids[i]]) != '0' or '1':
            continue
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])
        confidence = str(round(confidences[i], 2))
        color = colors[i]
        cv2.rectangle(img, (x, y), (x+w, y+h), color, 2)
        cv2.putText(img, f'{label}: {confidence}', (x, y+20), font, 2, (255, 255, 255), 2)

    cv2.imshow('Demo', img)
    
    # decrease the wait time/delay
    cv2.waitKey(1)
    
    # add a escape key to shutdown the program
```
